```python
import pandas as pd
import fractions
import random
data = {'prompts': [], 'response': []}
words = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
def decimal_to_word(decimal):
    if '.' not in str(decimal):
        return words[min(int(decimal), 9)] 
    whole_part, decimal_part = str(decimal).split('.')
    if int(whole_part) > 0:
        whole_word = words[min(int(whole_part), 9)] + " point "  
    else:
        whole_word = "zero point "
    decimal_word = ' '.join([words[min(int(digit), 9)] for digit in decimal_part])  
    
    return whole_word + decimal_word                          
def word_to_decimal(word_str):
    word_mapping = {
        'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
        'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9,
    }

    if 'point' not in word_str:
        return word_mapping[word_str]
    
    whole_part, fractional_part = word_str.split(' point ')
    fractional_value = sum(word_mapping[digit] * (0.1 ** (idx + 1)) for idx, digit in enumerate(fractional_part.split()))

    return word_mapping[whole_part] + fractional_value
def evaluate_equivalence(items):
    """Evaluate if all items in the list are equivalent in value."""
    values = []
    for item in items:
        if "%" in item:
            values.append(round(float(item.rstrip('%')) / 100, 5))
        elif "/" in item:
            frac = fractions.Fraction(item)
            values.append(round(float(frac), 5))
        elif item.replace('.', '', 1).isdigit():
            values.append(round(float(item), 5))
        else:
            values.append(round(word_to_decimal(item), 5))
    return len(set(values)) == 1
count = 0
while count < 10000:
    num_items = random.randint(2, 4) 
    items = []
    
    # 1. Select a random base value
    numerator = random.randint(1, 10)
    denominator = random.randint(1, 10)
    frac = fractions.Fraction(numerator, denominator)
    decimal_base = float(frac)
    
    # 2. Randomly generate first format of this base value
    format_type = random.choice(["percentage", "fraction", "decimal", "word"])
    if format_type == "percentage":
        items.append(f'{decimal_base * 100:.2f}%')
    elif format_type == "fraction":
        items.append(f'{frac}')
    elif format_type == "decimal":
        items.append(f'{decimal_base:.2f}')
    else: 
        items.append(decimal_to_word(decimal_base))
    
    # 3. Use the base value to generate the other items in different formats
    available_formats = ["percentage", "fraction", "decimal", "word"]
    available_formats.remove(format_type)
    
    for _ in range(num_items - 1):
        format_type = random.choice(available_formats)
        available_formats.remove(format_type)
        
        if format_type == "percentage":
            items.append(f'{decimal_base * 100:.2f}%')
        elif format_type == "fraction":
            items.append(f'{frac}')
        elif format_type == "decimal":
            items.append(f'{decimal_base:.2f}')
        else: 
            items.append(decimal_to_word(decimal_base))
    
    statement = '[' + ', '.join(['"' + item + '"' for item in items]) + ']'
    
    data['prompts'].append(statement)
    data['response'].append(evaluate_equivalence(items))
    count += 1

df = pd.DataFrame(data)
df2 = df.sample(50)
df2['prompts'] = df2['prompts'].str.replace('"', "'")
df2

```
